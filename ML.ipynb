2.2 k nearest neighbor (k-NN) classification


1. As a starter let run k-NN with k = 3 and evaluate its performance on Dv by the error rate


```python
from sklearn.neighbors import KNeighborsClassifier
# Fit the model on (Xa, Ya)
k = 3
clf = KNeighborsClassifier(n_neighbors = k)
clf.fit(Xa, Ya)
# Predict the labels of samples in Xv
Ypred_v = clf.predict(Xv)
# evaluate classification error rate
from sklearn.metrics import accuracy_score
error_v = 1-accuracy_score(Yv, Ypred_v)
```

2. Inspiring from that, train the k-nn classifier for different values of k ∈ N (in the range
[1, 40]) and evaluate its performance either on training set and validation set. Plot the
training and validation error curves as a function of k. From the plots, point out the overfitting issue.

```python
# some hints
k_vector = np.arange(1, 37, 2)   # define a vector of k = 1, 3, 5, ...
error_train = np.empty(k_vector.shape)
error_val = np.empty(k_vector.shape)

for ind, k in enumerate(k_vector):

    # fit with k
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(Xa, Ya)

    # ---------- PREDICT TRAIN SET ----------
    Ypred_train = clf.predict(Xa)
    error_train[ind] = 1 - accuracy_score(Ya, Ypred_train)

    # ---------- PREDICT VALIDATION SET ----------
    Ypred_val = clf.predict(Xv)
    error_val[ind] = 1 - accuracy_score(Yv, Ypred_val)


# --------- PLOT TRAIN + VALIDATION ERROR CURVES ----------
plt.figure(figsize=(8,5))
plt.plot(k_vector, error_train, label="training error")
plt.plot(k_vector, error_val, label="validation error")
plt.xlabel("k (number of neighbors)")
plt.ylabel("Error rate")
plt.title("k-NN error curves")
plt.legend()
plt.grid(True)
plt.show()
```

2. Replicate the experiments from section 2.2 with the normalized data and compare the
achieved performances with the one of section 2.2. Drawn the conclusions ?


```python
# Normalize the training data only
from sklearn.preprocessing import StandardScaler

sc = StandardScaler(with_mean=True, with_std=True)
sc.fit(Xa)      # fit only on training data

Xa_n = sc.transform(Xa)
Xv_n = sc.transform(Xv)

# Repeat k-NN experiments for normalized data
k_vector = np.arange(1, 37, 2)
error_train_n = np.empty(k_vector.shape)
error_val_n = np.empty(k_vector.shape)

for ind, k in enumerate(k_vector):

    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(Xa_n, Ya)

    # Training predictions
    Ypred_train_n = clf.predict(Xa_n)
    error_train_n[ind] = 1 - accuracy_score(Ya, Ypred_train_n)

    # Validation predictions
    Ypred_val_n = clf.predict(Xv_n)
    error_val_n[ind] = 1 - accuracy_score(Yv, Ypred_val_n)


# Plot curves
plt.figure(figsize=(8,5))
plt.plot(k_vector, error_train_n, label="training error (normalized)")
plt.plot(k_vector, error_val_n, label="validation error (normalized)")
plt.xlabel("k")
plt.ylabel("Error rate")
plt.title("k-NN error curves (normalized inputs)")
plt.legend()
plt.grid(True)
plt.show()
```

3. How to make the trained models less sensitive to the data split ?

To make the model less sensitive to a specific train/validation/test split, a robust approach is to use stratified k-fold cross-validation, ideally repeated several times. This technique provides a more stable and reliable estimate of performance by averaging results across multiple folds, which greatly reduces the variance caused by a single random partition. Additionally, increasing the dataset size, applying regularization, or selecting inherently more stable models can further mitigate this sensitivity.

```python
import matplotlib.pyplot as plt

# Table content
rows = [
    ["Stratified k-fold CV", "Preserves class proportions across folds",
     "More stable & reliable performance estimates"],

    ["Repeated CV", "Runs k-fold multiple times with different shuffles",
     "Further reduces variance from random splits"],

    ["More data", "Increase the size of the training set",
     "Improves generalization and reduces noise"],

    ["Regularization", "Apply L2, dropout, smoothing, etc.",
     "Prevents overfitting and stabilizes the model"],

    ["Stable models", "Choose models less sensitive to noise (e.g., Logistic Regression)",
     "Performance varies less across splits"],
]

col_labels = ["Method", "Description", "Benefit"]

# Plot
plt.figure(figsize=(12, 3))
plt.axis('off')

table = plt.table(
    cellText=rows,
    colLabels=col_labels,
    cellLoc='center',
    colColours=["#DCE6F1"] * len(col_labels),
    loc='center'
)

# Styling
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1, 1.4)

# Bold header
for key, cell in table.get_celld().items():
    r, c = key
    if r == 0:  # header row
        cell.set_text_props(weight='bold')
        cell.set_edgecolor("black")

plt.title("Techniques to Reduce Sensitivity to Data Splits", fontsize=14, pad=12)
plt.show()
```
